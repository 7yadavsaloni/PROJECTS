## Z TEST 

#### The mean value and standard deviation value for the Annual rainfall of the 'HARYANA DELHI & CHANDIGARH' state is taken from the dataset and than the sample is taken of years 1980-2015, its mean is calculated.
#### Now , we are supposed to see if the average Annual Rainfall is significantly lower than the value calculated before , taking type 1 Error to be 5%.

```python
#average annual rainfall in haryana
a4=ds[ds['SUBDIVISION']=='HARYANA DELHI & CHANDIGARH'].mean()['ANNUAL']
print('Avearge Annual Rainfall in HARYANA DELHI & CHANDIGARH :',a4)
```

```python
#population Standard deviation
sigmaP=ds[ds['SUBDIVISION']=='HARYANA DELHI & CHANDIGARH']['ANNUAL'].std()
print('Population Standard Deviation is :',sigmaP)
```

```python
#making HYPOTHESIS
print('Ho : mu =',a4)
print('Ha : mu <',a4)
```

```python
#taking the sample from year 1980-2015
a5=ds[(ds['SUBDIVISION']=='HARYANA DELHI & CHANDIGARH') & (ds['YEAR']>1980)].mean()['ANNUAL']
print('Mean of the Sample :',a5)
```

```python
#sample size
n1=ds[(ds['SUBDIVISION']=='HARYANA DELHI & CHANDIGARH') & (ds['YEAR']>1980)].count()['ANNUAL']
print('Sample size is :',n1)
```

```python
# need to check if we if we can accept the hypothesis at alpha = 0.05
Zc1= -1.96
Z1=(a5-a4)*(n1**0.5)/(sigmaP)
print('alpha = 0.05')
print('\nZ value is :',Z1)
if(Z1<Zc1):
    print("\nEnough evidence to reject null Hypothesis")
else:
    print("\nNot Enough Evidence to Reject the Null Hypothesis")
```

#### *Therefore, Null Hypothesis accepted as Z lies in non-rejection region*

## T TEST

#### The mean value for the Annual rainfall of the 'PUNJAB' state is taken from the dataset and than the sample is taken of years 2000-2015, its mean and Standard Deviation is calculated.
#### Now , we are supposed to see if the average Annual Rainfall is significantly lower than the value calculated before , taking type 1 Error to be 5%.

```python
#average annual rainfall in Punjab
a6=ds[ds['SUBDIVISION']=='PUNJAB'].mean()['ANNUAL']
print('Average Annual Rainfall in Punjab :',a6)
```

```python
#making HYPOTHESIS
print('Ho : mu =',a6)
print('Ha : mu <',a6)
```

```python
#taking the sample of Punjab from year 2000-2015
a7=ds[(ds['SUBDIVISION']=='PUNJAB') & (ds['YEAR']>=2000)].mean()['ANNUAL']
print('Mean of the Sample is :',a7)
```

```python
#Calculating the sample standard deviation since sigma is not given
sigmaS=ds[(ds['SUBDIVISION']=='PUNJAB') & (ds['YEAR']>=2000)].std()['ANNUAL']
print('Sample Standard Deviation is :',sigmaS)
```

```python
#sample size
n2=ds[(ds['SUBDIVISION']=='PUNJAB') & (ds['YEAR']>=2000)].count()['ANNUAL']
print('Sample size is :',n2)
```

```python
# need to check if we if we can accept the hypothesis at alpha = 0.05
tc1= -1.753 
df1=n2-1
t1=(a7-a6)*(n2**0.5)/(sigmaS)
print('alpha = 0.05')
print('\ndf =',df1)
print('\nt value is :',t1)
if(t1<tc1):
    print("\nEnough evidence to reject null Hypothesis")
else:
    print("\nNot Enough evidence to reject Null Hypothesis")
```

#### *Therefore, null hypothesis rejected as t lies in rejection region*

### Using Z Statistics

#### The population standard deviation is taken from the dataset and than the sample is taken of years 1980-2015, its mean Average Annual Rainfall is calculated.
#### We are supposed to constuct a 90% confidence interval for mu.

```python
n3=ds[(ds['SUBDIVISION']=='HARYANA DELHI & CHANDIGARH') & (ds['YEAR']>1980)].count()['ANNUAL']
print('Sample size is :',n3)
```

```python
#calculating the 90% CI 
Zc=1.645
muX1= a6 + ((Zc*sigma1)/(n3**0.5))
muI1= a6 - ((Zc*sigma1)/(n3**0.5))
mu1=[muI1,muX1]
print('CL 90%')
print('\nmu lies between :',mu1)
```
